## 상관관계 구하기
### 주어진 데이터에서 상관관계를 구하고, quality와의 상관관계가 가장 큰 값과, 가장 작은 값을 구한 다음 더하시오!
### 단, quality와 quality 상관관계 제외, 소수점 둘째 자리까지 출력

- 데이터셋 : ../input/red-wine-quality-cortez-et-al-2009/winequality-red.csv
- 오른쪽 상단 copy&edit 클릭 -> 예상문제 풀이 시작
- 스크립트 방식 권장: File -> Editor Type -> Script

```python
# import

import pandas as pd
import numpy as np

# 데이터 불러오기
df = pd.read_csv("data/winequality-red.csv")
```

```python
# eda

print(df.shape)
print(df.isna().sum())
df.head()
```

```python
# 상관관계 구하기.
df_corr = df.corr()
df_corr = df_corr[:-1]


#quality의 상관관계 중 큰값 작은값 차이 (아 ,, 이게 상관관계의 정도는 절대값으로 구해야겠구먼 ,,)
max_cor = abs(df_corr.quality).max()
min_cor = abs(df_corr.quality).min()

# 절대값으로 구한 상관관계 min, max의 부호 돌려놓기 

if max_cor not in df_corr.quality.values:
    max_cor = -max_corr
if min_cor not in df_corr.quality.values:
    min_cor = -min_corr

print(max_cor, min_cor)
print(round(max_cor+min_cor,2))
```
