44. data import

```python
import pandas as pd

DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv'
df = pd.read_csv(DataUrl)
df.head()
```

46. ⭐️ 데이터의 각 host_name의 빈도수를 구하고 빈도수 기준 내림차순 정렬한 데이터 프레임을 만들어라. 빈도수 컬럼은 counts로 명명하라


```python
# 내가 만든 답안
temp = df.host_name.value_counts().sort_values(ascending = False)
ans = pd.DataFrame(temp)
ans.columns = ['counts']
ans
```

```python
# 답안

ans = df.groupby('host_name').size().to_frame().rename(columns={0:'counts'})
ans.sort_values('counts',ascending=False) # 데이터 프레임 내에선 기준 열을 지정해야함. 
```

⭐️ 47. neighbourhood_group의 유일값으로 분류하여 각 그룹의 neighbourhood 개수를 구하라

```python
df.neighbourhood_group.unique()
```

```python
df.groupby(['neighbourhood_group','neighbourhood'], as_index=True).size()
# as_index = False 여야 ,, 데이터 프레임으로 만들어짐ㅊㅍ
```

```python
df.groupby(['neighbourhood_group','neighbourhood'], as_index=False).size()
# as_index = False 여야 ,, 데이터 프레임으로 만들어짐
```

⭐️ 48. neighbourhood_group의 값에 따른 neighbourhood컬럼 값 중 neighbourhood_group그룹의 최댓값들을 출력하라

```python
temp = df.groupby(['neighbourhood_group','neighbourhood'],as_index=False).size()
ans = temp.groupby('neighbourhood_group', as_index = False).max()
ans
```

⭐️ 49. neighbourhood_group 값에 따른 price값의 평균, 분산, 최대, 최소 값을 구하여라


그룹핑하고 어떤 값을 찾기 위해선 필요한 열만 먼저 추출하는게 좋을 것 같음.   
그리고 나서 문맥을 잘 파악하는거지!

```python
temp = df[['neighbourhood_group','price']]
temp.groupby('neighbourhood_group').agg(['mean','var','max','min'])
```

여기서 잘 파악할 것. 단순집계를 위해선 `agg`를 이용해야한다!

50. neighbourhood_group 값에 따른 reviews_per_month 평균, 분산, 최대, 최소 값을 구하여라


```python
temp = df[['neighbourhood_group','reviews_per_month']]
temp.groupby('neighbourhood_group').agg(['mean','var','max','min'])
```

51. neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균을 구하라


```python
temp = df[['neighbourhood', 'neighbourhood_group','price']] 
temp.groupby(['neighbourhood', 'neighbourhood_group']).mean()
```
