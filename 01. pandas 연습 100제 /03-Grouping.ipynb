⭐️ 52. neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균을 계층적 indexing 없이 구하라

계층적 인덱싱이란 위와 같이 차원이 나눠진, (2개의 그룹핑) * (price)이렇게 계층이 나뉜 것을 말함. (2개 그룹핑 된게 볼드 처리 되서 따로 빼지잖아, 그거 없이 하라는것!)

없으면 모든 칼럼이 같은 위상을 갖는것

```python
# as_index = False 는 계층적 인덱스 없이하는것!
temp = df[['neighbourhood', 'neighbourhood_group','price']] 
temp.groupby(['neighbourhood', 'neighbourhood_group'], as_index = False).mean()
```

```python
# 답안이라 하는데 왜 답안인지 모르겠음
Ans = df.groupby(['neighbourhood', 'neighbourhood_group']).price.mean().unstack()
Ans 
```

53. neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균을 계층적 indexing 없이 구하고 nan 값은 -999값으로 채워라


```python
# 결측값이 이상함. 결측값 채우는것은 .fillna(-999)로 하면 됨. 
temp = df[['neighbourhood', 'neighbourhood_group','price']] 
result = temp.groupby(['neighbourhood', 'neighbourhood_group'], as_index = False).mean()
result.price.isna().sum()
```

⭐️ 54. 데이터중 neighbourhood_group 값이 Queens값을 가지는 데이터들 중 neighbourhood 그룹별로 price값의 평균, 분산, 최대, 최소값을 구하라


```python
#groupby().price 이렇게 코드를 짜면 price가 집계항목이 된다. (따로 안빼도 됨)
df[df['neighbourhood_group'] == 'Queens'].groupby('neighbourhood').price.agg(['mean','var','max','min']).head()
```

⭐️ 55. 데이터중 neighbourhood_group 값에 따른 room_type 컬럼의 숫자를 구하고 neighbourhood_group 값을 기준으로 각 값의 비율을 구하여라

그니까 ,, 풀어서 설명하자면 

neighbourhood_group 값에 따른 room_type의 칼럼 수 A-1, A-2 이런식의 숫자 구함. 
neighbourhood_group의 값을 기준으로 비율을 구함 1/A, 2/A 이렇게 

unstack() # 인덱스를 칼럼으로 보냄. 따라서 데이터 프레임 차원을 재구조화하는 것. 

```python
# 나름 내가 만든 답안이야 ,, ㅋㅋ
temp1 = df[['neighbourhood_group','room_type']]
temp2 = temp1.groupby(['neighbourhood_group','room_type'],as_index= False).size()

lst1 = temp1.groupby('neighbourhood_group').size().to_list()
lst2 = []

for i in lst1 :
    lst2.append(i)
    lst2.append(i)
    lst2.append(i)

temp2['sum'] = lst2
temp2['ratio'] = temp2['size']/temp2['sum']
temp2
```
