⭐️ 24. 특정 칼럼의 특정 문자 제거 후 데이터 타입 바꾸어 저장  

str()은 데이터 타입을 문자열로 바꿔주는 함수이기도 하지만  
str[]로 사용할 시 문자열 슬라이싱 or 인덱싱으로 활용할 수 있다.

